# Modelling

## Content Based Filtering

### Prep before CBF

```python
# load data
tourismCBF = tourism_data.copy()
tourismCBF
```

```python
# Drop column that we dont use
dataCBF = tourismCBF.drop(['Description','Price','Rating','Coordinate', 'Lat', 'Long'],axis=1)
dataCBF
```

```python
# combine 2 variable, so tf-idf will have more keyword
datacc = dataCBF[['Category','City']].apply(lambda x : '{} {}'.format(x[0],x[1]), axis=1)
datacbf = pd.concat([dataCBF, datacc], axis = 1)
datacbf
```

```python
# matric to dense
tfidf_matrix.todense()
```

```python
# make df of matrix
cbf_tourism = pd.DataFrame(tfidf_matrix.todense(), 
                           columns=tf.get_feature_names(),
                           index=datacbf.Place_Name)
cbf_tourism
```

## Collaborative Filtering

### Prep before CF

```python
# define tourism rating to tourism rate variable
tourism_rate = rating.copy()
tourism_rate
```

User_id and Place_Id value already assign to sorted number from 1 to last, probably it seems we dont need to encode the data.  
We just need to change Place_ratings value into float.

But to keep the data safe and the stream didn't clash, let's just encode it. :-)

```python
# encoding using LabelEncoder sklearn
le = LabelEncoder()

tourism_rate['User_Id']=le.fit_transform(tourism_rate['User_Id'])
tourism_rate['Place_Id']=le.fit_transform(tourism_rate['Place_Id'])
tourism_rate
```

```python
# change Place_ratings value into float
tourism_rate['Place_Ratings'] = tourism_rate['Place_Ratings'].values.astype(np.float32)
tourism_rate
```

```python
# check data
len(tourism_rate['User_Id'].unique()), len(tourism_rate['Place_Id'].unique()) 
```

### Rating Normalization

```python
data_scale = tourism_rate.copy()
categorical_var=['User_Id', 'Place_Id']

feature_scale=[feature for feature in data_scale.columns if feature not in categorical_var]
scaler=MinMaxScaler()
scaler.fit(data_scale[feature_scale])
```

```python
scaled_data = pd.concat([data_scale[['User_Id', 'Place_Id']].reset_index(drop=True),
                         pd.DataFrame(scaler.transform(data_scale[feature_scale]), columns=feature_scale)],
                        axis=1)
scaled_data.head()
```

```python
# scramble data
datascr = scaled_data.sample(frac=1, random_state=42)
datascr
```

### Train test split

```python
# feature selection
X=datascr.drop(['Place_Ratings'],axis=1)
y=datascr.Place_Ratings

# set global random state
random_state = 42

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=random_state)
print("Input Training:",X_train.shape)
print("Input Test:",X_test.shape)
print("Output Training:",y_train.shape)
print("Output Test:",y_test.shape)
```

### Build Model

```python
# build class model
EMBEDDING_SIZE = 50

class RecommenderNet(keras.Model):
    
    def __init__(self, num_users, num_tourism, embedding_size, **kwargs):
        super(RecommenderNet, self).__init__(**kwargs)
        self.num_users = num_users
        self.num_tourism = num_tourism
        self.embedding_size = embedding_size
        self.user_embedding = layers.Embedding(
            num_users,
            embedding_size,
            embeddings_initializer="he_normal",
            embeddings_regularizer=keras.regularizers.l2(1e-6),
        )
        self.user_bias = layers.Embedding(num_users, 1)
        self.tourism_embedding = layers.Embedding(
            num_tourism,
            embedding_size,
            embeddings_initializer="he_normal",
            embeddings_regularizer=keras.regularizers.l2(1e-6),
        )
        self.tourism_bias = layers.Embedding(num_tourism, 1)

    def call(self, inputs):
        user_vector = self.user_embedding(inputs[:, 0])
        user_bias = self.user_bias(inputs[:, 0])
        tourism_vector = self.tourism_embedding(inputs[:, 1])
        tourism_bias = self.tourism_bias(inputs[:, 1])
        dot_user_tourism = tf.tensordot(user_vector, tourism_vector, 2)
        # Add all the components (including bias)
        x = dot_user_tourism + user_bias + tourism_bias
        # The sigmoid activation forces the rating to between 0 and 1
        return tf.nn.sigmoid(x)
```

```python
# compile model
num_users = len(tourism_rate['User_Id'].unique()) 
num_tourism = len(tourism_rate['Place_Id'].unique())

model = RecommenderNet(num_users, num_tourism, EMBEDDING_SIZE)
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
              optimizer=keras.optimizers.Adam(learning_rate=0.001),
              metrics=[tf.keras.metrics.RootMeanSquaredError()]
             )
```

```python
# define tensorboard_callback
logdir = os.path.join("logs", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"),'Adam')
tensorboard_callback = tf.keras.callbacks.TensorBoard(logdir, histogram_freq=1)
```

```python
history = model.fit(X_train,
                    y_train,
                    batch_size=64,
                    epochs=100,
                    verbose=1,
                    validation_data=(X_test, y_test),
                    callbacks=[tensorboard_callback]
                   )
```

## Make Recommendation System

### CBF System Recommendation

```python
def tourism_recommendations_cbf(tourism_name, similarity_data=cos_sim_df, items=datacbf[['Place_Name', 'Category', 'City']], k=5):
    """
    Tourism Recommendation base on similarity dataframe
 
    Parameter:
    ---
    tourism_name : type of data -> string (str)
                   Place Name (index similarity dataframe)
    similarity_data : type of data -> pd.DataFrame (object)
                      similarity dataframe, symmetric, with tourism as index and column
    items : type of data -> pd.DataFrame (object)
            Contains name and other feature which use to define similarity
    k : type of data -> integer (int)
        Count of recommendation that given
    ---
 
 
    in this index, we retrieve k with biggest similarity in matrix index that given (i).
    """
 
 
    # Retrieve data by using argpartition to partition indirectly along a given axis    
    # Dataframe change to numpy
    # Range(start, stop, step)
    index = similarity_data.loc[:,tourism_name].to_numpy().argpartition(
        range(-1, -k, -1))
    
    # retrieve data with biggest similarity from index
    closest = similarity_data.columns[index[-1:-(k+2):-1]]
    
    # Drop tourism name so the name that searching exclued from recommendations
    closest = closest.drop(tourism_name, errors='ignore')
 
    return pd.DataFrame(closest).merge(items).head(k)
```

```python
# retrieve tourism recommendations which similar with Farm House Susu Lembang
print("----" * 13)
print("Top 5 Recommendation")
print("----" * 13)
tourism_recommendations_cbf('Farm House Susu Lembang')
```

### CF System Recommendation

```python
# prepare data
tourismcf = tourism_data.copy()
tourismcf = tourismcf.drop(['Description','Coordinate', 'Lat', 'Long'],axis=1)
tourismcf.columns = ['tourism_id','tourism_name','category','city', 'rating','price']
ratingcf = rating.copy()

#dict of encode
user_ids = rating["User_Id"].unique().tolist()
user2user_encoded = {x: i for i, x in enumerate(user_ids)}
userencoded2user = {i: x for i, x in enumerate(user_ids)}
tourism_ids = rating["Place_Id"].unique().tolist()
tourism2tourism_encoded = {x: i for i, x in enumerate(tourism_ids)}
tourism_encoded2tourism = {i: x for i, x in enumerate(tourism_ids)}

# Let us get a user and see the top recommendations.
user_id = rating.User_Id.sample(1).iloc[0]
visited_by_user = rating[rating.User_Id == user_id]

# Find data that user didn't
tourism_not_visited = tourismcf[~tourismcf["tourism_id"].isin(visited_by_user.Place_Id.values)]["tourism_id"]
tourism_not_visited = list(set(tourism_not_visited)
                          .intersection(set(tourism2tourism_encoded.keys())))
tourism_not_visited = [[tourism2tourism_encoded.get(x)] for x in tourism_not_visited]
user_encoder = user2user_encoded.get(user_id)
user_tourism_array = np.hstack(
    ([[user_encoder]] * len(tourism_not_visited), tourism_not_visited))
```

```python
ratings = model.predict(user_tourism_array).flatten()
top_ratings_indices = ratings.argsort()[-5:][::-1]
recommended_tourism_ids = [
    tourism_encoded2tourism.get(tourism_not_visited[x][0]) for x in top_ratings_indices
]

print("Showing recommendations for user: {}".format(user_id))
print("====" * 9)
print("Movies with high ratings from user")
print("----" * 8)
top_tourism_user = (
    visited_by_user.sort_values(by="Place_Ratings", ascending=False)
    .head(5)
    .Place_Id.values
)
tourism_df_rows = tourismcf[tourismcf["tourism_id"].isin(top_tourism_user)]
for row in tourism_df_rows.itertuples():
    print(row.tourism_name, ":", row.category,":", row.city)

print("----" * 13)
print("Top 5 Recommendation")
print("----" * 13)
recommended_tourism = tourismcf[tourismcf["tourism_id"].isin(recommended_tourism_ids)]
for row in recommended_tourism.itertuples():
    print(row.tourism_name, ":", row.category, ":", row.city )
```
